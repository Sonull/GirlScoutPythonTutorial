# Load Sample Dataset

```python
data = pd.read_csv('https://raw.githubusercontent.com/jsiwu94/GirlScoutPythonTutorial/master/Dataset/countries.csv'
                   , sep=',')
```

```python
data.head(5)
```

# Compare the population growth in the US and China by Year

```python
us = data[data.country == 'United States']
```

```python
china = data[data.country == 'China']
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.title('Line Chart of US vs China Population by Year (in Million)')
plt.show()
```

```python
plt.scatter(us.year, us.population / 10**6)

plt.scatter(china.year, china.population / 10**6)
plt.xlabel('year')
plt.ylabel('population')
plt.title('Scatter Plot of US vs China Population by Year (in Million)')
```

## Bar Plot for Top 20 Most Populated Countries Over Time

```python
plt.style.use('seaborn-dark-palette')
data.groupby("country").population.sum().sort_values(ascending=False)[:20].plot.barh(align='center')
plt.title("Total Population by Country")
```

```python
plt.style.use('fivethirtyeight')
data.groupby("country").population.sum().sort_values(ascending=False)[:20].plot.bar()
plt.title("Total Population by Country")
```

# Putting it Altogether : Plot Top 4 Most Populated Countries by Year

```python
plt.style.use('ggplot')
plt.figure(figsize=(8,5))
plt.title('Top 4 Most Populated Countries Population by Year', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(data.year[data.country == 'China'], data.population[data.country == 'China'], 'b.-')
plt.plot(data.year[data.country== 'India'], data.population[data.country== 'India'], 'r.-')
plt.plot(data.year[data.country=='United States'], data.population[data.country=='United States'], 'g.-')
plt.plot(data.year[data.country== 'Indonesia'], data.population[data.country== 'Indonesia'], 'y.-')

plt.xticks(data.year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('Population')

plt.legend()


plt.show()
```
